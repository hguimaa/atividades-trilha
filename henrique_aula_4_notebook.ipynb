## Questão 2 📝

Crie uma classe `Produto` com os atributos `nome` e `preco`.

Implemente um método `aplicar_desconto(percentual)` que reduza o preço do produto conforme o percentual informado.

O método deve rejeitar percentuais fora do intervalo 0–100.

````python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco
    def aplicar_desconto(self, percentual):
        if percentual < 0 or percentual > 100:
            raise ValueError('O percentual deve estar entre 0 e 100')
        else:
            self.preco *= 1-(percentual/100)
            return self.preco
````

### 💡 Dica: 

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Valide o percentual antes de aplicar qualquer cálculo. Use `raise ValueError` para sinalizar entradas inválidas.*

## Questão 3 📝

Implemente uma classe `ContaBancaria` com operações de depósito, saque e consulta de saldo.

O saldo **não deve ser acessado diretamente** de fora da classe.

````python
class ContaBancaria:
    def __init__(self, nome, saldo):
        self.nome = nome
        self._saldo = saldo

    def deposito(self, quantidade):
        if quantidade < 0:
            raise ValueError('O depósito não pode ser negativo')
        
        self._saldo += quantidade
        return self._saldo

    def saque(self, quantidade):
        if quantidade > self._saldo:
            raise ValueError('O saque não pode ser maior que seu saldo atual.')
        
        self._saldo -= quantidade
        return self._saldo

    def getsaldo(self):
        return self._saldo
        
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Em Python, o prefixo `_` em um atributo (ex: `_saldo`) indica que ele é de uso interno — é uma convenção de encapsulamento.

Pense nas validações: faz sentido depositar um valor negativo? Sacar mais do que tem?

## Questão 4 📝

Crie uma classe `Livro` que controle o progresso de leitura.

A classe deve ter `titulo` e `paginas` (total), e um método `ler(qtd)` que avance as páginas lidas. Não permita que as páginas lidas ultrapassem o total.

Implemente também um método `progresso()` que retorne o percentual lido.

````python
class Livro:
    def __init__(self, titulo, total_paginas, paginas_lidas = 0):
        self.titulo = titulo
        self.total_paginas = total_paginas
        self.paginas_lidas = paginas_lidas

    def ler(self, qtd):
        if self.paginas_lidas + qtd > self.total_paginas:
            return 'A quantidade de páginas lidas ultrapassa o total de páginas.'
        else:
            self.paginas_lidas += qtd
            return self.paginas_lidas
        
    def progresso(self):
        percentual = (self.paginas_lidas*100) / self.total_paginas
        return f'Você leu {percentual}% do livro {self.titulo}!'
    
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Guarde as páginas já lidas como um atributo separado, iniciado em zero.

Porcentagem = `(paginas_lidas / paginas_total) * 100`.

## Questão 5 📝

Crie uma hierarquia de classes para funcionários.

- `Funcionario(nome, salario_base)` — método `calcular_salario()` retorna o salário base
- `Gerente` — herda de `Funcionario` e adiciona R$ 1.000 ao salário
- `Vendedor(nome, salario_base, vendas)` — herda de `Funcionario` e adiciona 10% do valor de vendas

Instancie um `Gerente` e um `Vendedor` e imprima o salário de cada um.

````python
class Funcionario:
    def __init__(self, nome, salario_base):
        self.nome = nome
        self.salario_base = salario_base
    
    def calcular_salario(self):
        return self.salario_base

class Gerente(Funcionario):
    def calcular_salario(self):
        return self.salario_base + 1000

class Vendedor(Funcionario):
    def __init__(self, nome, salario_base, vendas):
        super().__init__(nome, salario_base)
        self.vendas = vendas
    
    def calcular_salario(self):
        return self.salario_base + self.vendas*0.1


v1 = Vendedor('Bruno', 1000, 4000)
g1 = Gerente("Henrique", 30000)

print(v1.calcular_salario())
print(g1.calcular_salario())
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Herança: `class Filha(Pai)`. A subclasse pode **sobrescrever** métodos do pai simplesmente redefinindo-os.

Quando a subclasse tem atributos extras (como `vendas`), use `super().__init__(...)` para reaproveitar o `__init__` da classe pai.

## Questão 6 📝

Implemente um sistema de pagamentos usando polimorfismo.

Crie uma classe base `Pagamento` com um método `processar()`, e duas subclasses: `Pix` e `Cartao`, cada uma com sua própria implementação do método.

Ao final, crie uma lista com um objeto de cada tipo e chame `processar()` para todos.

````python
class Pagamento:
    def __init__(self, valor):
        self.valor = valor
    def processar(self):
        return f'Processando o pagamento de {self.valor}R$...'

class Pix(Pagamento):
    def __init__(self, valor):
        super().__init__(valor)

    def processar(self):
        return f'Processando o pagamento em PIX de {self.valor}R$...'

class Cartao(Pagamento):
    def __init__(self, valor, parcelas):
        super().__init__(valor)
        self.parcelas = parcelas
        
    def processar(self):
        return f'Processando o pagamento em CARTÃO de {self.valor}R$ dividido em {self.parcelas} parcelas...'
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Polimorfismo = mesma interface (`processar()`), comportamentos diferentes por classe.

Iterar sobre uma lista de objetos e chamar o mesmo método em cada um é a aplicação clássica — o Python resolve qual implementação usar automaticamente.

## Questão 7 📝

Crie uma classe `Carrinho` que permita:
- `adicionar_produto(nome, preco)` — adiciona um produto ao carrinho
- `remover_produto(nome)` — remove um produto pelo nome
- `total()` — retorna o valor total dos produtos no carrinho

````python
class Carrinho:
    def __init__(self, itens=None):
        self.itens = [] if itens is None else itens
    
    def adicionar_produto(self, nome, preco):
        self.itens.append({'nome' : nome, 'preco' : preco})
    
    def remover_produto(self, nome):
        self.itens = [item for item in self.itens if item['nome'] != nome]
    
    def total(self):
        return sum(item['preco'] for item in self.itens)
    

````

### 💡 Dica: 

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Use uma lista para armazenar os produtos. Tuplas `(nome, preco)` funcionam bem aqui.

`sum()` com um generator expression é uma forma elegante de somar os preços.
